```python
# if elif else
num1 = int(input("Enter the number: "))
num2 = 20
num3 = 30
if num1 > num2 :
  print("if part run")
elif num1 > num3 :
  print("elif part run")
else:
  print("else part run ")
```

```python
marks = int(input("Enter your marks: "))

if marks >= 95:
    print("Grade A")
elif marks >= 90:
    print("Grade B")
elif marks >= 75:
    print("Grade C")
elif marks >= 60:
    print("Grade D")
else:
    print("Fail")
```
